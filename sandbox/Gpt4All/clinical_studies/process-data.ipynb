```python
import os
from natsort import natsorted

directory = 'results/'

df = pd.DataFrame()
for file in natsorted(os.listdir(directory)):
    print(file)
    tdf = pd.read_excel(f'{directory}/{file}',engine='openpyxl').drop(columns='Unnamed: 0',axis=1)
    df = pd.concat([df,tdf]).reset_index(drop=True)

df[0:5]
```

```python
# ERROR: The prompt size exceeds the context window size and cannot be processed.
df = df[df['outcome_raw'].str.contains('ERROR: The prompt size exceeds the context window size and cannot be processed.')==False].reset_index(drop=True)
df = df[df['brand_name'].isnull()==False].reset_index(drop=True)
df = df[df['application_number'].isnull()==False].reset_index(drop=True)


print(len(df))
df[0:5]
```

```python
import re

def extract_metrics(text):
    """
    Splits the input text at every occurrence of the word "metric" followed by
    quotes and grabs the value wrapped in quotes immediately after it.

    Parameters:
    - text (str): The input text to search through.

    Returns:
    - list of str: A list containing the values found immediately after "metric".
    """
    # Adjusting our lens to focus on "metric": "value" patterns.
    # The pattern looks for the word "metric" followed by any amount of whitespace,
    # then a colon, possibly some more whitespace, and finally a quoted value.
    pattern = r'"metric"\s*:\s*"([^"]*)"'

    # Our detective work uses 'findall' to collect all matching quoted values.
    values = re.findall(pattern, text)

    return values

def extract_values(text):
    """
    Splits the input text at every occurrence of the word "metric" followed by
    quotes and grabs the value wrapped in quotes immediately after it.

    Parameters:
    - text (str): The input text to search through.

    Returns:
    - list of str: A list containing the values found immediately after "metric".
    """
    # Adjusting our lens to focus on "metric": "value" patterns.
    # The pattern looks for the word "metric" followed by any amount of whitespace,
    # then a colon, possibly some more whitespace, and finally a quoted value.
    pattern = r'"value"\s*:\s*"([^"]*)"'

    # Our detective work uses 'findall' to collect all matching quoted values.
    values = re.findall(pattern, text)

    return values

def extract_descriptors(text):
    """
    Splits the input text at every occurrence of the word "metric" followed by
    quotes and grabs the value wrapped in quotes immediately after it.

    Parameters:
    - text (str): The input text to search through.

    Returns:
    - list of str: A list containing the values found immediately after "metric".
    """
    # Adjusting our lens to focus on "metric": "value" patterns.
    # The pattern looks for the word "metric" followed by any amount of whitespace,
    # then a colon, possibly some more whitespace, and finally a quoted value.
    pattern = r'"descriptors"\s*:\s*"([^"]*)"'

    # Our detective work uses 'findall' to collect all matching quoted values.
    values = re.findall(pattern, text)

    return values

def equalize_list_lengths(list1, list2):
    """
    Checks the lengths of two lists and, if they differ, appends "N/A" to the shorter
    one until both lists have equal lengths.

    Parameters:
    - list1, list2 (list): The lists to be equalized.

    Modifies the input lists in place.
    """
    # First, let's see who's the shorter party here.
    length_difference = len(list1) - len(list2)

    # If list1 is the shorter one.
    if length_difference < 0:
        for _ in range(abs(length_difference)):
            list1.append("N/A")
    # Else if list2 is the shorter one.
    elif length_difference > 0:
        for _ in range(length_difference):
            list2.append("N/A")
    # If they're already equal, they both leave the courtroom with dignity.

```

```python
# Datafame Format 1: (combined statements)
statements = []
brands = []
apps = []
texts = []

position = 0
for outcome in df['outcome_raw']:

    # if position == 250: # TODO: handle no valid value extracted
    #     continue

    brand_name = df['brand_name'][position]
    app_no = df['application_number'][position]
    original_text = df['clinical_studies'][position]

    metrics = extract_metrics(outcome)
    values = extract_values(outcome)
    descriptors = extract_descriptors(outcome)

    equalize_list_lengths(metrics, values)
    equalize_list_lengths(metrics, descriptors)

    for i in range(len(metrics)):
        statement = f'{metrics[i]}: {values[i]} ({descriptors[i]})'

        statements.append(statement)
        brands.append(brand_name)
        apps.append(app_no)
        texts.append(original_text)

    position += 1

data_format1 = pd.DataFrame().assign(brand_name=brands,application=apps,original_text=texts,statement=statements)
data_format1
```

```python
# Datafame Format 2: (separated metrics, values, descriptors)
metric_list = []
value_list = []
descriptor_list = []

texts = []
brands = []
apps = []

position = 0
for outcome in df['outcome_raw']:

    # if position == 250:
    #     continue

    brand_name = df['brand_name'][position]
    app_no = df['application_number'][position]
    original_text = df['clinical_studies'][position]

    metrics = extract_metrics(outcome)
    values = extract_values(outcome)
    descriptors = extract_descriptors(outcome)

    equalize_list_lengths(metrics, values)
    equalize_list_lengths(metrics, descriptors)

    for i in range(len(metrics)):
        metric_list.append(metrics[i])
        value_list.append(values[i])
        descriptor_list.append(descriptors[i])

        texts.append(original_text)
        brands.append(brand_name)
        apps.append(app_no)

    position += 1

data_format2 = pd.DataFrame().assign(brand_name=brands,application=apps,original_text=texts,metrics=metric_list,values=value_list,descriptors=descriptor_list)
data_format2
```
